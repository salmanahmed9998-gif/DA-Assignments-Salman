```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
file_path = '/content/drive/MyDrive/Salaries.csv'
df = pd.read_csv(file_path)
```

Drop rows where EmployeeName and JobTitle are 'Not provided'

```python
df = df.dropna(subset=['EmployeeName', 'JobTitle'])
```

Replace 'Not Provided' with 0 in BasePay OvertimePay	OtherPay	Benefits

```python
df['BasePay'] = df['BasePay'].replace('Not Provided', 0)
df['OvertimePay'] = df['OvertimePay'].replace('Not Provided', 0)
df['OtherPay'] = df['OtherPay'].replace('Not Provided', 0)
df['Benefits'] = df['Benefits'].replace('Not Provided', 0)
```

Change BasePay	OvertimePay	OtherPay	Benefits type to float

```python
df['BasePay'] = df['BasePay'].astype(float)
df['OvertimePay'] = df['OvertimePay'].astype(float)
df['OtherPay'] = df['OtherPay'].astype(float)
df['Benefits'] = df['Benefits'].astype(float)
```

Display all employees name and basepay columns who have a BasePay greater than 200,000.

```python
high_basepay = df[df['BasePay'] > 200000]
print("Employees with BasePay > 200,000:\n", high_basepay)
```

Find all employees whose JobTitle contains the word “Manager” (case = False)

```python
managers = df[df['JobTitle'].str.contains('Manager', case=False)]
print("\nEmployees with 'Manager' in JobTitle:\n", managers)
```

Retrieve only the columns: EmployeeName, JobTitle, and TotalPay.

```python
employee_info = df[['EmployeeName', 'JobTitle', 'TotalPay']]
print("\nEmployee Info (EmployeeName, JobTitle, TotalPay):\n", employee_info)
```

Find count of employees who have zero OvertimePay.

```python
zero_overtime = len(df[df['OvertimePay'] == 0])
print("\nCount of employees with zero OvertimePay:", zero_overtime)
```

Calculate the TotalPay for each JobTitle where totalpay > 300000

```python
high_totalpay = df[df['TotalPay'] > 30000].groupby('JobTitle')['TotalPay'].sum()
print("\nTotalPay for each JobTitle where TotalPay > 30,000:\n", high_totalpay)
```

Find the top 5 JobTitles with the highest average TotalPayBenefits show in descending order

```python
top_5_jobtitles = df.groupby('JobTitle')[['TotalPay', 'Benefits']].mean().sort_values(by='TotalPay', ascending=False).head(5)
print("\nTop 5 JobTitles with highest average TotalPayBenefits:\n", top_5_jobtitles)
```

Replace all missing Benefits values with 0

```python
df['Benefits'] = df['Benefits'].fillna(0)
print("\nMissing Benefits replaced with 0")
```

Identify the top employee name by TotalPay in each Year. Show year, employeename and totalpay dataframe

```python
top_employee = df.groupby('Year')['TotalPay'].idxmax()
top_employee_info = df.loc[top_employee][['Year', 'EmployeeName', 'TotalPay']]
print("\nTop employee by TotalPay in each Year:\n", top_employee_info)
```

Find the number of employees per JobTitle.

```python
employee_count_by_title = df['JobTitle'].value_counts()
print("\nNumber of employees per JobTitle:\n", employee_count_by_title)
```
